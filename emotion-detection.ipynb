```python
test = pd.read_csv('/kaggle/input/emotion-classification-nlp/emotion-labels-test.csv')
```

```python
val = pd.read_csv('/kaggle/input/emotion-classification-nlp/emotion-labels-val.csv')
```

```python
from sklearn.model_selection import train_test_split
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.naive_bayes import MultinomialNB
from sklearn.metrics import accuracy_score, classification_report
```

```python
X_train, y_train = train['text'], train['label']
X_test, y_test = test['text'], test['label']
X_val, y_val = val['text'], val['label']
```

```python
vectorizer = CountVectorizer()
X_train_vec = vectorizer.fit_transform(X_train)
X_val_vec = vectorizer.transform(X_val)
X_test_vec = vectorizer.transform(X_test)
```

```python
classifier = MultinomialNB()
classifier.fit(X_train_vec, y_train)
```

```python
predictions_val = classifier.predict(X_val_vec)
```

```python
accuracy_val = accuracy_score(y_val, predictions_val)
report_val = classification_report(y_val, predictions_val)
```

```python
print(f"Validation Accuracy: {accuracy_val}")
print("Validation Classification Report:\n", report_val)
```

```python
predictions_test = classifier.predict(X_test_vec)
```

```python
accuracy_test = accuracy_score(y_test, predictions_test)
report_test = classification_report(y_test, predictions_test)
```

```python
print(f"\nTest Accuracy: {accuracy_test}")
print("Test Classification Report:\n", report_test)
```

```python
from sklearn.model_selection import GridSearchCV

param_grid = {'C': [0.001, 0.01, 0.1, 1, 10, 100]}
grid_search = GridSearchCV(LogisticRegression(max_iter=1000), param_grid, cv=5)
grid_search.fit(X_train_vec, y_train)

best_params = grid_search.best_params_
best_model = grid_search.best_estimator_

# Evaluate the best model on validation and test sets
best_val_predictions = best_model.predict(X_val_vec)
best_test_predictions = best_model.predict(X_test_vec)

val_accuracy = accuracy_score(y_val, best_val_predictions)
test_accuracy = accuracy_score(y_test, best_test_predictions)

val_report = classification_report(y_val, best_val_predictions)
test_report = classification_report(y_test, best_test_predictions)

print(f"Best Model Parameters: {best_params}")
print(f"Best Model Validation Accuracy: {val_accuracy}")
print("Best Model Validation Classification Report:\n", val_report)
print(f"Best Model Test Accuracy: {test_accuracy}")
print("Best Model Test Classification Report:\n", test_report)
```
